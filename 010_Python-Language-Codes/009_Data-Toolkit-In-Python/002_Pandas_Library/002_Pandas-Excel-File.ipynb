# Working with `.xlsx` Files in Pandas

## 1. Reading Excel Files
- `df = pd.read_excel("file.xlsx")` → Reads an Excel file into a DataFrame.
- `df = pd.read_excel("file.xlsx", sheet_name="Sheet1")` → Reads a specific sheet.
- `df = pd.read_excel("file.xlsx", usecols="A:C")` → Reads only specific columns.
- `df = pd.read_excel("file.xlsx", skiprows=3)` → Skips the first 3 rows.
- `df = pd.read_excel("file.xlsx", nrows=10)` → Reads only the first 10 rows.

## 2. Writing to Excel Files
- `df.to_excel("output.xlsx")` → Saves DataFrame to an Excel file.
- `df.to_excel("output.xlsx", index=False)` → Saves without the index column.
- `df.to_excel("output.xlsx", sheet_name="NewSheet")` → Saves to a specific sheet.
- `df.to_excel("output.xlsx", columns=["col1", "col2"])` → Writes only selected columns.

## 3. Working with Multiple Sheets
- `xls = pd.ExcelFile("file.xlsx")` → Loads an Excel file with multiple sheets.
- `df1 = pd.read_excel(xls, sheet_name="Sheet1")` → Reads a specific sheet.
- `df_dict = pd.read_excel("file.xlsx", sheet_name=None)` → Reads all sheets into a dictionary of DataFrames.

## 4. Appending & Merging Data
- `df_existing = pd.read_excel("file.xlsx")`
- `df_combined = pd.concat([df_existing, df_new], ignore_index=True)` → Merges two DataFrames.

## 5. Handling Missing Values in Excel Files
- `df.isnull().sum()` → Counts missing values.
- `df.fillna(value)` → Replaces missing values.
- `df.dropna()` → Removes rows with missing values.

## 6. Checking Excel File Properties
- `df.shape` → Returns `(rows, columns)`.
- `df.dtypes` → Displays data types of all columns.
- `df.describe()` → Provides statistical summary.
